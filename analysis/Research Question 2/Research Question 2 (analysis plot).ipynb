## Below scripts helps to displays average Initial response time for 59 countries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mode


merge = pd.read_csv('new_rq2merged.csv')
merge = merge.dropna()

average_initial_response_time_x = merge.groupby('Country')['Initial Response_x'].mean()
average_initial_response_time_y = merge.groupby('Country')['Initial Response_y'].mean()


average_initial_response_time_x_sorted = average_initial_response_time_x.sort_values(ascending=False)
average_initial_response_time_y_sorted = average_initial_response_time_y.sort_values(ascending=False)

common_countries = list(set(average_initial_response_time_x_sorted.index).intersection(average_initial_response_time_y_sorted.index))

average_initial_response_time_x_sorted = average_initial_response_time_x_sorted.loc[common_countries]
average_initial_response_time_y_sorted = average_initial_response_time_y_sorted.loc[common_countries]

mean_values_x = average_initial_response_time_x_sorted.values
mean_values_y = average_initial_response_time_y_sorted.values

mean_x = average_initial_response_time_x_sorted.mean()
median_x = average_initial_response_time_x_sorted.median()
mode_x = mode(average_initial_response_time_x_sorted)[0][0]

mean_y = average_initial_response_time_y_sorted.mean()
median_y = average_initial_response_time_y_sorted.median()
mode_y = mode(average_initial_response_time_y_sorted)[0][0]

fig, ax = plt.subplots(figsize=(30, 10))
width = 0.40

x = np.arange(len(common_countries))
x_sorted = sorted(range(len(average_initial_response_time_x_sorted)), key=lambda i: average_initial_response_time_x_sorted.iloc[i], reverse=True)
rects1 = ax.bar(x - width/2, average_initial_response_time_x_sorted.values[x_sorted], width, edgecolor='black', color='#cbf7f4', label='F-PKI with WebAssembly Average Initial Response Time')
rects2 = ax.bar(x + width/2, average_initial_response_time_y_sorted.values[x_sorted], width, edgecolor='black', color='#f9c5c8', label='Without F-PKI Average Initial Response Time')

for i, rect in enumerate(rects1):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height/2, f'{mean_values_x[x_sorted[i]]:.2f} ms', ha='center', va='center', rotation='vertical')

for i, rect in enumerate(rects2):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height/2, f'{mean_values_y[x_sorted[i]]:.2f} ms', ha='center', va='center', rotation='vertical')

ax.text(0.25, 0.92, f"F-PKI, Total 59 Countries Average Initial Response Time Mean: {mean_x:.2f} ms", transform=ax.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.25, 0.85, f"F-PKI, Total 59 Countries Average Initial Response Time Median: {median_x:.2f} ms", transform=ax.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.25, 0.78, f"F-PKI, Total 59 Countries Average Initial Response Time Mode: {mode_x:.2f} ms", transform=ax.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))


ax.text(0.51, 0.92, f"Without F-PKI, Total 59 Countries Average Initial Response Time Mean: {mean_y:.2f} ms", transform=ax.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.51, 0.85, f"Without F-PKI, Total 59 Countries Average Initial Response Time Median: {median_y:.2f} ms", transform=ax.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.51, 0.78, f"Without F-PKI, Total 59 Countries Average Initial Response Time Mode: {mode_y:.2f} ms", transform=ax.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))

ax.set_xticks(x)
ax.set_xticklabels(average_initial_response_time_x_sorted.index[x_sorted], rotation=90)

ax.set_ylabel('Average Initial Response Time (ms)')
ax.set_xlabel('Countries')

ax.legend()

plt.tight_layout()
#plt.savefig("rq2country.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mode

merge = pd.read_csv('new_rq2merged.csv')
merge = merge.dropna()

average_initial_response_time_x = merge.groupby('Website')['Initial Response_x'].mean()
average_initial_response_time_y = merge.groupby('Website')['Initial Response_y'].mean()

average_initial_response_time_x_sorted = average_initial_response_time_x.sort_values(ascending=False)
average_initial_response_time_y_sorted = average_initial_response_time_y.sort_values(ascending=False)

common_websites = list(set(average_initial_response_time_x_sorted.index).intersection(average_initial_response_time_y_sorted.index))

average_initial_response_time_x_sorted = average_initial_response_time_x_sorted.loc[common_websites]
average_initial_response_time_y_sorted = average_initial_response_time_y_sorted.loc[common_websites]

mean_values_x = average_initial_response_time_x_sorted.values
mean_values_y = average_initial_response_time_y_sorted.values

mean_x = average_initial_response_time_x_sorted.mean()
median_x = average_initial_response_time_x_sorted.median()
mode_x = mode(average_initial_response_time_x_sorted)[0][0]

mean_y = average_initial_response_time_y_sorted.mean()
median_y = average_initial_response_time_y_sorted.median()
mode_y = mode(average_initial_response_time_y_sorted)[0][0]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(40, 30))
width = 0.50

x = np.arange(len(common_websites))
x_sorted = sorted(range(len(average_initial_response_time_x_sorted)), key=lambda i: average_initial_response_time_x_sorted.iloc[i], reverse=True)
rects1 = ax1.bar(x, average_initial_response_time_x_sorted.values[x_sorted], width, edgecolor='black', color='#d2b4de', label='F-PKI with WebAssembly Average Initial Response Time')
ax1.set_ylabel('Average Initial Response Time (ms)')
ax1.set_title('Average Initial Response Time for F-PKI with WebAssembly')

for i, rect in enumerate(rects1):
    height = rect.get_height()
    ax1.text(rect.get_x() + rect.get_width() / 2, height + 5, f' {mean_values_x[x_sorted[i]]:.2f} ms', ha='center', va='bottom', rotation='vertical')

ax1.text(0.25, 0.92, f"F-PKI, Total 92 Websites Average Initial Response Time Mean: {mean_x:.2f} ms", transform=ax1.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax1.text(0.25, 0.85, f"F-PKI, Total 92 Websites Average Initial Response Time Median: {median_x:.2f} ms", transform=ax1.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax1.text(0.25, 0.78, f"F-PKI, Total 92 Websites Average Initial Response Time Mode: {mode_x:.2f} ms", transform=ax1.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))

rects2 = ax2.bar(x, average_initial_response_time_y_sorted.values[x_sorted], width, edgecolor='black', color='#f8f9c5', label='Without F-PKI Average Initial Response Time')
ax2.set_ylabel('Average Initial Response Time (ms)')
ax2.set_title('Average Initial Response Time for Without F-PKI')

for i, rect in enumerate(rects2):
    height = rect.get_height()
    ax2.text(rect.get_x() + rect.get_width() / 2, height + 5, f' {mean_values_y[x_sorted[i]]:.2f} ms', ha='center', va='bottom', rotation='vertical')

ax2.text(0.25, 0.92, f"Without F-PKI, Total 92 Websites Average Initial Response Time Mean: {mean_y:.2f} ms", transform=ax2.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax2.text(0.25, 0.85, f"Without F-PKI, Total 92 Websites Average Initial Response Time Median: {median_y:.2f} ms", transform=ax2.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax2.text(0.25, 0.78, f"Without F-PKI, Total 92 Websites Average Initial Response Time Mode: {mode_y:.2f} ms", transform=ax2.transAxes,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))


ax1.set_xticks(x)
ax1.set_xticklabels(average_initial_response_time_x_sorted.index[x_sorted], rotation=90)
ax2.set_xticks(x)
ax2.set_xticklabels(average_initial_response_time_x_sorted.index[x_sorted], rotation=90)
ax1.set_xlabel('Websites')
ax2.set_xlabel('Websites')
ax1.legend()
ax2.legend()


plt.tight_layout()
#plt.savefig("rq2websites.png", dpi=300, bbox_inches='tight')
plt.show()
```
