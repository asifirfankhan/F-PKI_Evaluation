## Below script helps to show from 94 websites how many unique websites show Case 2 behaviour

```python
import pandas as pd


df3 = pd.read_csv('new_case2_no_rtt_size_duration.csv')
unique_websites = df3['Website'].unique()


num_unique_websites = len(unique_websites)


print("Unique Websites:")
for website in unique_websites:
    print(website)
print(f"Total Unique Websites: {num_unique_websites}")
```

## Below script merge the clean First phase (new_case2_no_rtt_size_duration) csv file (that shows case 2 behaviour) with Third phase csv. Third phase csv file contain without F-PKI dataset

```python
import pandas as pd


df1 = pd.read_csv('new_case2_no_rtt_size_duration.csv')
df2 = pd.read_csv('Third phase.csv')


merged_df = pd.merge(df1, df2, on=['Website', 'Country'])
merged_df.to_csv('nowasmrttsizes10merged.csv', index=False)
```
