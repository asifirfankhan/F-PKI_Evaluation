## Below scripts helps to displays out of 48 websites, how many websites from a specific country, the F-PKI map server provide website certificate data log JSON file that does not have RTT, size, and duration values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random


merge = pd.read_csv('new_case2_no_rtt_size_duration.csv')
merge = merge.dropna()


country_counts = merge['Country'].value_counts()
country_counts_sorted = country_counts.sort_values(ascending=False)


fig, ax = plt.subplots(figsize=(12, 6))  
colors = [plt.cm.Set3(random.randint(0, 9)) for _ in range(len(country_counts_sorted))]
country_counts_sorted.plot(kind='bar', ax=ax, edgecolor='black',color=colors, width=0.8) 



for i, count in enumerate(country_counts_sorted):
    ax.text(i, count, str(count), ha='center', va='bottom')


ax.set_xlabel('Country')
ax.set_ylabel('Number of Websites')
ax.set_xticklabels(country_counts_sorted.index, rotation=90)




plt.tight_layout()
#plt.savefig("case2country.png", dpi=300, bbox_inches='tight') 
plt.show()
```

## Below scripts helps to displays total number of times each website’s certificate data log is provided by the F-PKI map server, which does not contain any data about the RTT, size, and duration values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random


merge = pd.read_csv('new_case2_no_rtt_size_duration.csv')
merge = merge.dropna()


country_counts = merge['Website'].value_counts()
country_counts_sorted = country_counts.sort_values(ascending=False)


fig, ax = plt.subplots(figsize=(12, 6))  
colors = [plt.cm.Set3(random.randint(0, 9)) for _ in range(len(country_counts_sorted))]
country_counts_sorted.plot(kind='bar', ax=ax, edgecolor='black',color=colors, width=0.8) 



for i, count in enumerate(country_counts_sorted):
    ax.text(i, count, str(count), ha='center', va='bottom')


ax.set_xlabel('Websites')
ax.set_ylabel('Number of Websites')
ax.set_xticklabels(country_counts_sorted.index, rotation=90)




plt.tight_layout()
#plt.savefig("case2website.png", dpi=300, bbox_inches='tight') 
plt.show()
```

## Below script check the average initial response time of 48 websites using F-PKI and without F-PKI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mode


merge = pd.read_csv('nowasmrttsizes10merged.csv')
merge = merge.dropna()


average_initial_response_time_x = merge.groupby('Website')['Initial Response_x'].mean()
average_initial_response_time_y = merge.groupby('Website')['Initial Response_y'].mean()


average_initial_response_time_x_sorted = average_initial_response_time_x.sort_values(ascending=False)
average_initial_response_time_y_sorted = average_initial_response_time_y.sort_values(ascending=False)


mean_with_fpki = average_initial_response_time_x_sorted.mean()
median_with_fpki = average_initial_response_time_x_sorted.median()
mode_with_fpki = mode(average_initial_response_time_x_sorted)[0][0]

mean_without_fpki = average_initial_response_time_y_sorted.mean()
median_without_fpki = average_initial_response_time_y_sorted.median()
mode_without_fpki = mode(average_initial_response_time_y_sorted)[0][0]


fig, ax = plt.subplots(figsize=(20, 10))
width = 0.50

x = np.arange(len(average_initial_response_time_x_sorted)) 
ax.bar(x - width/2, average_initial_response_time_x_sorted, width, edgecolor='black',color = '#3498db',label= 'F-PKI with WebAssembly Average Initial Response Time (case 2)')
ax.bar(x + width/2, average_initial_response_time_y_sorted, width, edgecolor='black',color = '#abebc6', label='Without F-PKI Average Initial Response Time (case 2)')

ax.text(0.15, 0.92, f"F-PKI, Total 48 websites Average Initial Response Time Mean: {mean_with_fpki:.2f} ms", transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.15, 0.85, f"F-PKI, Total 48 websites Average Initial Response Time Median: {median_with_fpki:.2f} ms", transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.15, 0.78, f"F-PKI, Total 48 websites Average Initial Response Time Mode: {mode_with_fpki:.2f} ms", transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))

ax.text(0.43, 0.92, f"Without F-PKI, Total 48 websites Average Initial Response Time Mean: {mean_without_fpki:.2f} ms", transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.43, 0.85, f"Without F-PKI, Total 48 websites Average Initial Response Time Median: {median_without_fpki:.2f} ms", transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))
ax.text(0.43, 0.78, f"Without F-PKI, Total 48 websites Average Initial Response Time Mode: {mode_without_fpki:.2f} ms", transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))

ax.set_xticks(x)
ax.set_xticklabels(average_initial_response_time_x_sorted.index, rotation=90)

ax.set_ylabel('Average Initial Response Time')
ax.set_xlabel('Websites')

ax.legend()

plt.tight_layout()
#plt.savefig("case2InitialResponse.png", dpi=300, bbox_inches='tight') 
plt.show()
```
