## Below scripts helps to displays total number of times each website waits for the F-PKI map server to provide a website certificate data log

```python
import pandas as pd
import matplotlib.pyplot as plt
import random


merge = pd.read_csv('new_waiting_time_exceed.csv')

country_counts = merge['Website'].value_counts()
#As "microsoft.com" shows security warning we remove the website from data set
country_counts_sorted = country_counts[country_counts.index != 'microsoft.com'].sort_values(ascending=False)


fig, ax = plt.subplots(figsize=(12, 6))  
colors = [plt.cm.Set3(random.randint(0, 9)) for _ in range(len(country_counts_sorted))]
country_counts_sorted.plot(kind='bar', ax=ax, edgecolor='black', color=colors, width=0.8) 

for i, count in enumerate(country_counts_sorted):
    ax.text(i, count, str(count), ha='center', va='bottom')

ax.set_xlabel('Websites')
ax.set_ylabel('Count')

ax.set_xticklabels(country_counts_sorted.index, rotation=90)

plt.tight_layout()
#plt.savefig("case4website.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Below scripts helps to displays out of 78 websites how many websites from a specific country, the F-PKI map server is not able to provide a website certificate data log.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random


merge = pd.read_csv('new_waiting_time_exceed.csv')
merge = merge.dropna()


country_counts = merge['Country'].value_counts()
country_counts_sorted = country_counts.sort_values(ascending=False)


fig, ax = plt.subplots(figsize=(12, 6))  
colors = [plt.cm.Set3(random.randint(0, 9)) for _ in range(len(country_counts_sorted))]
country_counts_sorted.plot(kind='bar', ax=ax, edgecolor='black',color=colors, width=0.8) 



for i, count in enumerate(country_counts_sorted):
    ax.text(i, count, str(count), ha='center', va='bottom')


ax.set_xlabel('Country')
ax.set_ylabel('Number of Websites')
ax.set_xticklabels(country_counts_sorted.index, rotation=90)




plt.tight_layout()
#plt.savefig("case4awebsite.png", dpi=300, bbox_inches='tight') 
plt.show()
```
